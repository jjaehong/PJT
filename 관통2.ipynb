3. 데이터 전처리(정제)

- 데이터 전처리 단계 분석을 진행하기 전 데이터를 정제,가공하는 단계

- 다음과 같은 과정 포함
    
    1. 불완전하거나 오류가 있는 데이터를 제거하여 데이터의 품질을 개선
    
    2. 중복 데이터 제거

    3. 분석하기 적절한 형식으로 데이터 변환

- 데이터 전처리 및 분석에 사용되는 파이썬 패키지 :

    1. Numpy 

    - 수학 계산용 패키지

    - 다차원 배열을 빠르게 사용할 수 있도록 지원하는 파이썬 패키지
    
    2. Pandas

    - 원하는 데이터만 추출하거나 데이터를 분석할 때 활용되는 패키지

    - 마치 엑셀을 다루듯 고성능의 데이터 구조를 만들 수 있음
    
    - Series(1차원 배열)과 DataFrame(2차원 배열)이라는 효율적인(편리하게 해줌) 자료구조 제공

    3. Matplotlib : 데이터 시각화

    - 다양한 종류의 그래프와 도표를 생성하고 데이터를 시각적으로 표현 가능
