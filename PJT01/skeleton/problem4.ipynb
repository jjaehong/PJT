# 문제4. 상품과 옵션 정보들을 담고 있는 새로운 객체를 만들어 반환하시오.
- [힌트] 상품 리스트와 옵션 리스트를 금융상품 코드를 기준으로 매칭할 수 있습니다.
- [힌트] 아래와 같은 순서로 데이터를 출력하며 진행합니다.
- 1. 응답을 json 형식으로 변환합니다.
- 2. key 값이 "result" 인 데이터를 변수에 저장합니다.
- 3. 2번의 결과 중 key 값이 "baseList" 인 데이터를 변수에 저장합니다.
- 4. 2번의 결과 중 key 값이 "optionList" 인 데이터를 변수에 저장합니다.
- 5. 3번에서 저장된 변수를 순회하며, 4번에서 저장된 값들에서 금융 상품 코드가 
-     같은 모든 데이터들을 가져와 새로운 딕셔너리로 저장합니다.
-     저장 시, 명세서에 맞게 출력되도록 저장합니다.
- 6. 5번에서 만든 딕셔너리를 결과 리스트에 추가합니다.

```python
def get_deposit_products():
    # 본인의 API KEY 로 수정합니다.
    my_api_key = "c0aecf40a0bc63eecd50da8c11f5902d"

    # API 요청 URL
    my_url = f'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json?auth={my_api_key}&topFinGrpNo=020000&pageNo=1'

    # api 요청 보내기
    my_response = requests.get(my_url).json()

    # 원하는 데이터 출력
    deposit_products_baselist = my_response['result']['baseList']
    deposit_product_option_list = my_response['result']['optionList']

    # 상품과 옵션의 정보를 담은 새로운 리스트 작성
    result = []
    for j in deposit_products_baselist:
        new_dict = {}
        new_dict["금융회사명"] = j['kor_co_nm']
        new_dict["금융상품명"] = j['fin_prdt_nm']     
        intr_info = new_dict["금리정보"] = []
    
        for i in deposit_product_option_list:
            if i['fin_prdt_cd'] == j["fin_prdt_cd"]:
                old_dict = {}
                old_dict["저축 금리"] = i['intr_rate']
                old_dict["최고 우대금리"] = i['intr_rate2']
                old_dict["저축 금리 유형"] = i['intr_rate_type']
                old_dict["저축 금리 유형명"] = i['intr_rate_type_nm']
                old_dict["저축 기간"] = i['save_trm']
                intr_info.append(old_dict)
        result.append(new_dict)
    
    return result
```

```python
# 아래 코드는 수정하지 않습니다.
if __name__ == '__main__':
    # json 형태의 데이터 반환
    my_result = get_deposit_products()
    # prrint.prrint(): json 을 보기 좋은 형식으로 출력
    pprint.pprint(my_result)
```
