# 문제3. 응답 중 정기예금 상품들의 옵션 리스트를 출력하도록 구성합니다.
##  이 때, 원하는 데이터만 추출하여 새로운 리스트를 만들어 반환하는 함수를 작성하시오.
- [힌트] 아래와 같은 순서로 데이터를 출력하며 진행합니다.
- 1. 응답을 json 형식으로 변환합니다.
- 2. key 값이 "result" 인 데이터를 출력합니다.
- 3. 위의 결과 중 key 값이 "optionList" 인 데이터를 변수에 저장합니다.
- 4. 3번에서 저장된 값을 반복하며, 원하는 데이터만 추출 및 가공하여 결과 리스트에 저장합니다.

```python
def get_deposit_products():
    # 본인의 API KEY 로 수정합니다.
    my_api_key = "c0aecf40a0bc63eecd50da8c11f5902d"
    
    # API 요청 URL
    my_url = f'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json?auth={my_api_key}&topFinGrpNo=020000&pageNo=1'
   
    # api 요청 보내기
    my_response = requests.get(my_url).json()
    
    # 전체 정기예금 옵션 리스트 출력
    deposit_product_option_list = my_response['result']['optionList']

    # 옵션을 담은 새로운 리스트 작성
    result = []
    for i in deposit_product_option_list:
        new_dict = {}
        new_dict["금융상품코드"] = i['fin_prdt_cd']
        new_dict["저축 금리"] = i['intr_rate']
        new_dict["최고 우대금리"] = i['intr_rate2']
        new_dict["저축 금리 유형"] = i['intr_rate_type']
        new_dict["저축 금리 유형명"] = i['intr_rate_type_nm']
        new_dict["저축 기간"] = i['save_trm']
        result.append(new_dict)

    return result
```

```python
# 아래 코드는 수정하지 않습니다.
if __name__ == '__main__':
    # json 형태의 데이터 반환
    my_result = get_deposit_products()
    # prrint.prrint(): json 을 보기 좋은 형식으로 출력
    pprint.pprint(my_result)
```
